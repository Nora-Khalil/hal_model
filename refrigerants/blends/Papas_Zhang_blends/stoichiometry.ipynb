```python
# script to get stoichiometric mole fractions of Papas blends 
```

```python
import numpy as np
```

```python
def get_stoic(CH2F2_frac,C3H2F4_frac):
    CH2F2 = CH2F2_frac
    C3H2F4 = C3H2F4_frac
    C = 1*CH2F2 + 3*C3H2F4
    H = 2*CH2F2 + 2*C3H2F4
    F = 2*CH2F2 + 4*C3H2F4

    if F > H: 
        HF = H
    if F < H: 
        HF = F
    if F==H: 
        HF = F
    if F-HF>0: 
        CF2O = (F-HF)/2
    if F-HF==0: 
        CF2O = 0 
    if C-CF2O>0: 
        CO2 = C-CF2O
    if C-CF2O==0:
        CO2 = 0
    if H-HF>0:
        H2O = (H-HF)/2
    if H-HF==0: 
        H2O = 0 

    O = CO2*2 + H2O*1 + CF2O*1

    O2 = O/2
    N2 = 3.76 * O2 



    stoich_dictionary={'CH2F2(38)':CH2F2/O2, 'C3H2F4(1)': C3H2F4/O2, 'O2(3)': O2/O2, 'N2': N2/O2}


    RHS_C = 1*CH2F2 + 3*C3H2F4
    RHS_H = 2*CH2F2 + 2*C3H2F4
    RHS_F = 2*CH2F2 + 4*C3H2F4
    RHS_O = 2*O2
    RHS_N = 3.76*2*O2

    LHS_C = 1*CO2 + 1*CF2O  
    LHS_H = 1*HF + 2*H2O 
    LHS_F = 1*HF + 2*CF2O
    LHS_O = 2*CO2 + 1*H2O + 1*CF2O
    LHS_N = N2*2

    assert(RHS_C==LHS_C)
    assert(RHS_H==LHS_H)
    assert(RHS_F==LHS_F)
    assert(RHS_O==LHS_O)
    assert(RHS_N==LHS_N)
    
    print(CH2F2_frac, C3H2F4_frac)
    print(stoich_dictionary)
    print()
    
    return stoich_dictionary
```

```python
tetra_mole_fraction= np.linspace(0,1,25)
array_of_1 = np.array([1 for x in tetra_mole_fraction])
di_mole_fraction = array_of_1 - tetra_mole_fraction
```

```python
list_of_dictionaries = []

for d,t in zip(di_mole_fraction,tetra_mole_fraction):
    stoich_dictionary = get_stoic(d,t)
    list_of_dictionaries.append((t,stoich_dictionary))
#1.0 CH2F2 + 0.0 C3H2F4 + 1.0(O2 + 3.76 N2)<=> 1.0 CO2 + 2.0 HF + 0 H2O + 0 CF2O + 3.76 N2
```

```python
for fraction, dictionary in list_of_dictionaries:
    print(fraction, dictionary)
```

```python
tetra_mole_fraction= np.linspace(0,1,25)
array_of_1 = np.array([1 for x in tetra_mole_fraction])
propane_mole_fraction = array_of_1 - tetra_mole_fraction
```

```python
list_of_dictionaries = []

for d,t in zip(propane_mole_fraction,tetra_mole_fraction):
    stoich_dictionary = get_stoic_t_p(d,t)
    list_of_dictionaries.append((t,stoich_dictionary))
```
