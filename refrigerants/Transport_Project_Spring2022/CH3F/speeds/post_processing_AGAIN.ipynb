```python
import os
import re
import matplotlib.pylab as plt


directory = "."

list_of_files = os.listdir(directory)
list_of_txt_files = [file for file in list_of_files if re.match('speed[0-9]+\.txt', file)] #list of only .txt files 

speeds_dictionary = {}
for txt_file in list_of_txt_files:
    match = re.match('speed0([0-9]+)\.txt', txt_file)
    with open(txt_file,'r') as f:
        read = f.readlines()
        for i in read:
            tokens = i.split('/n')
            flamespeed = float(tokens[0])
            speeds_dictionary[int(match.group(1))] = flamespeed
        
print(speeds_dictionary.items())
    
```

```python
#flamespeeds = speeds_dictionary.values()
#species = speeds_dictionary.keys()


fig = plt.figure()
ax1 = fig.add_subplot(111)

for species, flamespeed in speeds_dictionary.items():
    ax1.plot(species, flamespeed, '*', c='b', markersize=14)
    
plt.rcParams["figure.figsize"] = (25,10)

plt.xlabel('Species Identification Number')
plt.ylabel('Burning Velocity (m/s)')
plt.title("Burning Velocities For Each 'Erroneous' Species")

plt.rcParams.update({'font.size': 30})



'''
for x in speeds_dictionary.values():
    flamespeeds.append(x)
for x in speeds_dictionary.keys():
    species.append(x)    
'''


```
